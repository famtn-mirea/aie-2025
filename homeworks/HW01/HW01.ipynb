```python
import numpy as np
```

#Примеры создания массивов

Из обычных списков

```python
list_array = np.array([1, 2, 3]) # Одномерный массив
list_array
```

```python
list_array = np.array([[1, 2, 3], [4, 5, 6]]) # Двумерный массив
list_array
```

С использованием zeros

```python
zeros_array = np.zeros(5) # Одномерный массив из пяти нулей
zeros_array
```

```python
zeros_array = np.zeros((2, 3)) # Двумерный массив из нулей формы 2 на 3
zeros_array
```

С использованием ones

```python
ones_array = np.ones(5) # Одномерный массив из пяти единиц
ones_array
```

```python
ones_array = np.ones((3, 5)) # Двумерный массив из нулей формы 3 на 5
ones_array
```

С использованием arange

```python
arr = np.arange(5)
arr
```

```python
arr = np.arange(2, 8) # С началом и концом
arr
```

```python
arr = np.arange(0, 10, 2) # С шагом
arr
```

# Операции с массивом

```python
#Поэлементое сложение
np.arange(1, 7) + np.arange(-1, -7, -1)
```

```python
#Поэлементое умножение
np.arange(1, 7) * np.arange(-1, -7, -1)
```

```python
# Поэлементное умножение двумерного массива

np.array([[1, 2], [3, 4]]) * np.array([[2, 2], [2, 2]])
```

```python
# Матричное умножение

np.array([[1, 2], [3, 4]]) @ np.array([[2, 2], [2, 2]])
```

# Простая статистика

```python
arr = np.random.randn(100000) # 100000 значений из стандартного нормального распределения (мат ожидание 0, стандартное отклонение 1)
arr[:100] # Первые 100
```

```python
arr.mean() # Среднее значение
```

```python
arr.std() # Стандартное отклонение
```
